```python
# signal brut
rec = recordings[0][3]
brut = rec
plt.plot(brut)
plt.show()
```

```python
# signal decimated by factor of 10
resamp = signal.resample(brut, round(len(brut)/2))
#resamp = resamp[350:700]
plt.plot(resamp)
plt.show()
```

```python
x = brut  # Slection du canal 1
rate = 4000
# Cration de instants d'chantillons
t = np.linspace(0, x.shape[0]/rate, x.shape[0])

# Trac du signal
plt.plot(t, x, label="Signal chantillonn")
plt.grid()
plt.xlabel(r"$t$ (s)")
plt.ylabel(r"Amplitude")
plt.title(r"Signal sonore")
plt.show()
```

```python
# Signal after low pass filter ( fc = 170 )
fc = 170 # Cut-off frequency of the filter
fs = 4000
w = fc / (fs / 2) # Normalize the frequency
b, a = signal.butter(3, w, 'low')
f = signal.filtfilt(b, a, resamp)
plt.plot(f)
plt.show()
```

```python
peaks, _ = find_peaks(f, height=(np.amax(f)/2))
plt.plot(f)
plt.plot(peaks, f[peaks], "x")
plt.plot(np.zeros_like(f), "--", color="gray")
plt.show()
```

```python
# Normalised signal
normalized_vector = f / np.linalg.norm(f)
plt.plot(normalized_vector)
plt.show()
```
